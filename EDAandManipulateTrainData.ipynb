```python
def frame_to_dict (x): 
    all_list = list(x)
    all_dict = {}
    for thing in all_list:
        all_dict[thing] = all_list.count(thing)
    return all_dict
```

```python
def condense_name (x, L, other = "Other"):
    to_condense = list(x)
    for i in range(len(to_condense)):
        if to_condense[i] not in L:
            to_condense[i] = other
    return to_condense

# test = train_df["BsmtExposure"]
# print(frame_to_dict(test))
# print(frame_to_dict(condense_name(test, ["Gd", "Mn"])))

# test_2 = [1, 2, 3, 4]
# print(condense_name(test_2, [2, 3]))

#print(frame_to_dict(condense_name(train_df["BsmtQual"], ["Ex", "TA"], "Something")))

train_df.BldgType = condense_name(train_df["BldgType"], ["1Fam"], "Other")
train_df.BsmtExposure = condense_name(train_df["BsmtExposure"], ["Gd", "None"], "NoGd")
train_df.BsmtQual = condense_name(train_df["BsmtQual"], ["Ex", "None"], "Other")
train_df.Condition1 = condense_name(train_df["Condition1"], ["Norm", "RRAn"], "Other")
train_df.Exterior1st = condense_name(train_df["Exterior1st"], ["AsbShng", "BrkFace"], "Other")
train_df.Foundation = condense_name(train_df["Foundation"], ["BrkTil", "Slab"], "Other")
train_df.HeatingQC = condense_name(train_df["HeatingQC"], ["Ex"], "NotEx")
train_df.MSZoning = condense_name(train_df["MSZoning"], ["RL", "C (all)"], "Other")
train_df.SaleCondition = condense_name(train_df["SaleCondition"], ["Normal"], "Other")
train_df.GarageType = condense_name(train_df["GarageType"], ["Attchd", "None"], "Other")

train_df.loc[ train_df.OverallQual == 1, 'OverallQual'] = 2
train_df.loc[ train_df.OverallCond == 9, 'OverallCond'] = 5
train_df.loc[ train_df.OverallCond == 7, 'OverallCond'] = 6
train_df.loc[ train_df.OverallCond == 8, 'OverallCond'] = 6
train_df.loc[ train_df.OverallCond == 1, 'OverallCond'] = 2
train_df.loc[ train_df.OverallCond == 2, 'OverallCond'] = 3
train_df.loc[ train_df.RoofMatl == 'Metal', 'RoofMatl'] = 'Other'
train_df.loc[ train_df.RoofMatl == 'WdShake', 'RoofMatl'] = 'Other'
train_df.loc[ train_df.RoofMatl == 'Membran', 'RoofMatl'] = 'Other'
train_df.loc[ train_df.RoofMatl == 'Tar&Grv', 'RoofMatl'] = 'Other'
train_df.loc[ train_df.RoofMatl == 'Roll', 'RoofMatl'] = 'Other'
train_df.loc[ train_df.RoofMatl == 'ClyTile', 'RoofMatl'] = 'Other'
train_df.loc[ train_df.RoofMatl == 'CompShg', 'RoofMatl'] = 'Other'
train_df.loc[ train_df.ExterCond == 'Po', 'ExterCond'] = 'Fa'
train_df.loc[ train_df.ExterCond == 'TA', 'ExterCond'] = 'Gd'
train_df.loc[ train_df.ExterCond == 'Ex', 'ExterCond'] = 'Gd'
train_df.loc[ train_df.Foundation == 'Wood', 'Foundation'] = 'Other'
train_df.loc[ train_df.Foundation == 'Stone', 'Foundation'] = 'Other'

#bsmt_has_bath_list = [bsmt_has_bath_list[x] - bsmt_has_bath_list[x] + 1 if bsmt_has_bath_list[x] >= 1 else bsmt_has_bath_list[x] * 0 for x in range(len(bsmt_has_bath_list))]
```
