# Multiple Linear Regression Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import linear_model
```

```python
pd.set_option('display.max_columns', 170)
```

```python
data = pd.read_csv("compiled_features.csv") # loading the comobined csv
data.head() # checking te first 5 rows of the data frame
```

```python
# seperating predictors and targets
x_value = data.drop('LogSalePrice', axis = 1)
y_value = data.LogSalePrice
```

```python
# seperating training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_value, y_value, test_size=0.3, random_state=42)
```

```python
# getting the score, RSS & intercept of train & test data
lm = linear_model.LinearRegression()
lm.fit(X_train, y_train)
print("R^2 for train set: %f" %lm.score(X_train, y_train)) # coefficient of determination
print('RSS for train set: %.3f' %(np.sum((y_train - lm.predict(X_train))**2))) # residual sum of square
print('The intercept is %.3f' %(lm.intercept_)) # y-intercept

print('-'*50)

print("R^2 for test  set: %f" %lm.score(X_test, y_test)) # coefficient of determination
print('RSS for train set: %.3f' %(np.sum((y_test - lm.predict(X_test))**2))) # residual sum of square
```

```python
pd.set_option('display.max_rows', 170)
colnames = x_value.columns
result = pd.DataFrame(lm.coef_).transpose()
result.columns = colnames.tolist()
result['intercept'] = lm.intercept_ 
result = result.transpose()
result.columns = ['coefficient']
result.sort_values(by = 'coefficient', ascending = False)
```

```python
print('negative slope column names: \n %s' %(x_value.columns[lm.coef_< -0.5]))
print('*'*100)
print('positive slope column names: \n %s' %(x_value.columns[lm.coef_> 0.5]))
```
