```python
import sklearn
import warnings
import numpy as np
import sklearn.model_selection as ms
from sklearn import linear_model
import pandas as pd
data = pd.read_csv('compiled_features.csv')
```

```python
pd.set_option('display.max_columns',None)
#data = data.drop('Unnamed: 0', axis=1)
data
```

```python
target = data['LogSalePrice']
features = data.drop(columns = ['LogSalePrice'])
```

```python
# Trying MLR with latest data

from sklearn import linear_model
ols = linear_model.LinearRegression()

x_m = features
y_m = target

ols.fit(x_m, y_m)

print("beta_1, beta_2: " + str(np.round(ols.coef_, 3)))
print("beta_0: " + str(np.round(ols.intercept_, 3)))
print("RSS: %.2f" % np.sum((ols.predict(x_m) - y_m) ** 2))
print("R^2: %.5f" % ols.score(x_m, y_m))
```

```python
#Trying Lasso with latest data

from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn import datasets


lasso  = Lasso()
alphas = np.arange(.00001, .001, .0001)
lasso.set_params(normalize=True)
coefs_lasso  = []

for alpha in alphas:
        lasso.set_params(alpha=alpha)
        lasso.fit(features, target)  
        coefs_lasso.append(lasso.coef_)

coefs_lasso = pd.DataFrame(coefs_lasso, index = alphas, columns = features.columns)  
coefs_lasso
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import datasets
param_grid = [{'alpha':list(np.arange(.00001,.01,.00001))}]

para_search = GridSearchCV(estimator = lasso, param_grid = param_grid, scoring = 'explained_variance', return_train_score = True)
para_search = para_search.fit(features,target)            
               
```
